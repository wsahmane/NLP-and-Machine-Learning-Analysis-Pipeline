```python
reviews = pd.read_csv("IDTATextPreprocessing.csv")
```

```python
reviews.head()
```

```python
del reviews["Unnamed: 0"]
```

```python
reviews.head()
```

```python
#change the text column datatype to string
reviews = reviews.astype({'Review':'string'})
```

```python
#get the review text for preprocessing
text = reviews['Review']
```

```python
data = reviews.Review.values.tolist()
```

```python
data
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence)))

words = list(sent_to_words(data))
```

```python
print(words[:1][0][:30])
```

```python
import gensim.corpora as corpora

# Create Dictionary
id2word = corpora.Dictionary(words)

# Create Corpus
texts = words

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1][0][:30])
```

```python
from pprint import pprint

# number of topics
num_topics = 10

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)

# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
#doc_lda = lda_model[corpus]
```
