```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
census = pd.read_csv("census_catgr.csv")
```

```python
census.head()
```

```python
census['No of hours']=census['No of hours'].fillna(0)
```

```python
#transform numerical attributes into categorical ones
census['No of hours'] = pd.qcut(census['No of hours'], q=4, duplicates='drop')
census['No of hours'].value_counts(sort=False)
```

```python
census.head()
```

```python
census.shape
```

```python
census1 = census.sample(frac=0.01, replace=False, random_state=1)
census1.shape
```

```python
data_list = []
for row in range(0,5697):
    data_list.append([str(census1.values[row,column]) for column in range(0, 19)])
    
rules = apriori(data_list, min_support=0.25, min_confidence=0.25, min_length=2)
results = list(rules)
```

```python
print(results)
```

```python
#Put the results into a dataframe to have the rules listed in a table
df = pd.DataFrame(columns=('Items','Antecedent','Consequent','Support','Confidence','Lift'))

Support =[]
Confidence = []
Lift = []
Items = []
Antecedent = []
Consequent=[]

for RelationRecord in results:
    for ordered_stat in RelationRecord.ordered_statistics:
        Support.append(RelationRecord.support)
        Items.append(RelationRecord.items)
        Antecedent.append(ordered_stat.items_base)
        Consequent.append(ordered_stat.items_add)
        Confidence.append(ordered_stat.confidence)
        Lift.append(ordered_stat.lift)

df['Items'] = list(map(set, Items))                                   
df['Antecedent'] = list(map(set, Antecedent))
df['Consequent'] = list(map(set, Consequent))
df['Support'] = Support
df['Confidence'] = Confidence
df['Lift']= Lift

#Sort resulted dataframe by Lift
df.sort_values(by ='Lift', ascending = False, inplace = True)
```

```python
 df.reset_index(drop=True, inplace=True)
```
